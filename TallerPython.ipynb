```python
# Uso del bucle for
for i in range(5):
    if i % 2 == 0:
        print(str(i) + " es par")
    else:
        print(str(i) + " es impar")
```

```python
# Listas, un objeto importante
pares = []
impares = []

# Uso del bucle while
i = 0
while True:
    
    if i % 2 == 0:
        pares.append(i)
    else:
        impares.append(i)
    i += 1
    
    if i == 10:
        break

print("Número pares: " + str(pares))
print("Número impares: " + str(impares))
```

```python
# Funciones
def sep_par_impar(n, pares_impares):
    """
    Esta funcion separa pares e impares
    
    :param n: 
    :return: Lista con pares o impares
    """
    # Diccionarios, otro objeto clave en python
    resultado = {"pares": [], "impares": []}
    
    for j in range(n):
        if j % 2 == 0:
            resultado["pares"].append(j)
        else:
            resultado["impares"].append(j)
    
    if pares_impares == "pares" or pares_impares == "impares":
        return resultado[pares_impares]
    else:
        return "Error, elija si desea ver pares o impares"


print("Pares:\n"
      + str(sep_par_impar(20, "pares")) + "\n")
print("Impares:\n"
      + str(sep_par_impar(20, "impares")) + "\n")
print(sep_par_impar(20, "par"))
```

```python
# Objetos
class Cliente:
    def __init__(self, id_c, dinero_disp):
        self.id_c = id_c
        self.dinero_disp = dinero_disp
    
    def compra(self, cant_compra, tienda):
        """
        Cliente compra producto de tienda
        
        :type tienda: Tienda
        """
        gasto = cant_compra*tienda.precio_prod
        
        if cant_compra <= tienda.cant_prducto \
                and gasto <= self.dinero_disp:
            
            self.dinero_disp -= gasto
            tienda.cant_prducto -= cant_compra
            tienda.ingresos += gasto
            
            print("Cliente: " + str(self.id_c)
                  + "\nTipo: Normal"
                  + "\nCantidad comprada: " + str(cant_compra)
                  + "\nGasto (Bs): " + str(gasto)
                  + "\nDinero disponible (Bs): " + str(self.dinero_disp)
                  + "\n")
        else:
            print("Cliente: " + str(self.id_c)
                  + "\nNo logró comprar nada\n")


class ClientePremium(Cliente):
    def __init__(self, id, dinero_disp, descuento):
        super(ClientePremium, self).__init__(id, dinero_disp)
        self.descuento = descuento
    
    def compra(self, cant_compra, tienda):
        """
        Cliente compra producto de tienda
        
        :type tienda: Tienda
        """
        gasto = cant_compra*tienda.precio_prod
        desc = round(gasto*self.descuento, 0)
        gasto_ef = gasto - desc
        
        if cant_compra <= tienda.cant_prducto \
                and gasto_ef <= self.dinero_disp:
            
            self.dinero_disp -= gasto_ef
            tienda.cant_prducto -= cant_compra
            tienda.ingresos += gasto_ef
            
            print("Cliente: " + str(self.id_c)
                  + "\nTipo: Premium"
                  + "\nCantidad comprada: " + str(cant_compra)
                  + "\nGasto sin dscto. (Bs): " + str(gasto)
                  + "\nDescuento (Bs): " + str(desc)
                  + "\nGasto con dscto. (Bs): " + str(gasto_ef)
                  + "\nDinero disponible (Bs): " + str(self.dinero_disp)
                  + "\n")
        else:
            print("Cliente: " + str(self.id_c)
                  + "\nNo logró comprar nada\n")
    
    
class Tienda:    
    def __init__(self, nombre, cant_producto, precio_prod):
        self.nombre = nombre
        self.cant_prducto = cant_producto
        self.precio_prod = precio_prod
        self.ingresos = 0
    
    def ver_estado(self):
        print("Tienda: " + self.nombre
              + "\nPrecio unitario (Bs): " + str(self.precio_prod)
              + "\nCant. Producto: " + str(self.cant_prducto)
              + "\nIngresos (Bs): " + str(self.ingresos)
              + "\n")
```

```python
tienda_a = Tienda("Dña. Juanita", 10, 2)
pedro = Cliente(1, 10)
juan = ClientePremium(2, 20, 0.1)

tienda_a.ver_estado()
```

```python
pedro.compra(2, tienda_a)
tienda_a.ver_estado()
juan.compra(8, tienda_a)
tienda_a.ver_estado()
pedro.compra(2, tienda_a)
tienda_a.ver_estado()
juan.compra(8,tienda_a)
```

```python
import random
import matplotlib.pyplot as plt


class Entorno:

    def __init__(self):
        self.tienda = Tienda("Dña. Juanita", 100, 2)
        self.tiempo_atencion = 8*60
        self.clientes = []
        self.hist_hora_visitas = []
        self.hist_ingresos = []
        self.hist_inventario = []
    
    @staticmethod
    def gen_tell():
        tell = random.triangular(5,30,10)
        return round(tell, 0)
    
    def simular(self):
        tiempo = 0
        id_c = 0
        while tiempo <= self.tiempo_atencion:
            
            tiempo += self.gen_tell()
            id_c += 1 
            if random.random() <= 0.5:
                dinero_disp = round(random.uniform(10,30),0)
                self.clientes.append(Cliente(id_c,
                                             dinero_disp))
            else:
                dinero_disp = round(random.uniform(20,40),0)
                desc = round(random.uniform(0.05,0.20),2)
                self.clientes.append(ClientePremium(id_c,
                                                    dinero_disp,
                                                    desc))
            print("Hora: " + str(tiempo))
            cant_compra = round(random.triangular(5,20,10),0)
            self.clientes[-1].compra(cant_compra, 
                                     self.tienda)
            self.tienda.ver_estado()
            
            # Guardar
            self.hist_hora_visitas.append(tiempo)
            self.hist_ingresos.append(self.tienda.ingresos)
            self.hist_inventario.append(self.tienda.cant_prducto)
    
    def ver_ingresos(self):
        plt.plot(self.hist_hora_visitas,self.hist_ingresos,"-")
        plt.ylabel("Ingresos (Bs)")
        plt.title("Evolución de ingresos en el día")
        plt.show()
    
    def ver_inventario(self):
        plt.plot(self.hist_hora_visitas,self.hist_inventario,"-")
        plt.ylabel("Inventario (Unid.)")
        plt.title("Evolución de inventario en el día")
        plt.show()
```

```python
entorno_sim = Entorno()
entorno_sim.simular()
```

```python
entorno_sim.ver_ingresos()
entorno_sim.ver_inventario()
```

## 2. Algebra Lineal con Python

En esta sección usaremos python para realizar algunas operaciones fundamentales del algebra lineal. Es importante notar que estas operaciones son muy importantes para el desarrollo de distintos modelos estadísticos.

```python
import numpy as np

# Vectores

a = np.array([1,2,3])
print("a = \n"+str(a))
print("Dimensión: " + str(a.shape) 
      + "\nTipo:" + str(type(a))
      + "\n")

b = np.array([[2],[4],[3]])
print("b = \n"+str(b))
print("Dimensión: " + str(b.shape)
      + "\nTipo:" + str(type(a))
      + "\n")
```

```python
# Matrices

A = np.array([[1, 2, 3], [4, 5, 6], [1, 4, 3]])
print("A = \n"+str(A))
print("Dimensión: " + str(A.shape) 
      + "\nTipo:" + str(type(A))
      + "\n")

B = np.array([[5, 4, 2], [6, 7, 2], [2, 1, 3]])
print("B = \n"+str(B))
print("Dimensión: " + str(B.shape) 
      + "\nTipo:" + str(type(B))
      + "\n")
```
